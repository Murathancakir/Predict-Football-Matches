```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

import warnings
warnings.filterwarnings("ignore")
```

# All Leagues 2023

```python
# Read data from csv, excel
data = pd.read_csv(r"C:\Workspace\PredictFootballMatches\PredictSuperLeague\csv\full_dataset.csv")

# Get 2023 Data
data_2023 = data[data["Season"] == 2223]

# Create Features
conditions = [(data_2023.FTR == "H"),(data_2023.FTR == "A"),(data_2023.FTR == "D")]
values = [1,2,0]
data_2023['MatchResult'] = np.select(conditions, values, default=data_2023['FTR'])

# Calculate Parameter
data_2023.reset_index(inplace=True,drop=True)

for ind in data_2023.index:

    if data_2023.loc[ind]["MatchResult"] == 1:
        data_2023.at[ind,"WinningBet"] = data_2023.at[ind,"PSH"]

    if data_2023.loc[ind]["MatchResult"] == 2:
        data_2023.at[ind,"WinningBet"] = data_2023.at[ind,"PSA"]

    if data_2023.loc[ind]["MatchResult"] == 0:
        data_2023.at[ind,"WinningBet"] = data_2023.at[ind,"PSD"]

# Get Count Statistics
countMatchResult = Counter(data_2023.MatchResult)
countWinningBet = Counter(data_2023.WinningBet)

resultValues = countMatchResult.values()
mylabels = ["HomeTeam", "AwayTeam", "Draw"]

plt.title("Match Result Percentage")
plt.pie(resultValues, labels = mylabels, autopct='%1.1f%%')

n_bins = 20
fig, axs = plt.subplots(1)
plt.title("Winning Bet Distribution")
axs.hist(countWinningBet, bins=n_bins)
plt.show()
```
